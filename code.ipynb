```python
import csv
import random

class Book:
    def __init__(self, book_id, name, author, language, category, average_rate, maturity, publisher, published_date, page_count):
        self.book_id = book_id
        self.name = name
        self.author = author
        self.language = language
        self.category = category
        self.average_rate = average_rate
        self.maturity = maturity
        self.publisher = publisher
        self.published_date = published_date
        self.page_count = page_count

def create_books_from_csv(file_path):
    books = []

    with open(file_path, 'r', encoding='utf-8') as file:
        reader = csv.reader(file)
        next(reader)  
        for row in reader:
            book_id = row[0]
            name = row[1]
            author = row[2]
            language = row[3]
            category = row[4]
            average_rate = float(row[5])
            maturity = int(row[6])
            publisher = row[7]
            published_date = float(row[8])
            page_count = float(row[9])

            # Create a book object and add it to the list
            book = Book(book_id, name, author, language, category, average_rate, maturity, publisher, published_date, page_count)
            books.append(book)

    return books

# Define the file path and name
file_path = r'C:\Users\Lenovo\Desktop\learnin\CPI-CS\algo\recommandation system\google_books_dataset.csv'
# Call the function to create books from the CSV file
books = create_books_from_csv(file_path)

# Iterate over the books and print their attributes
for book in books:
    print(f"ID: {book.book_id}")
    print(f"Name: {book.name}")
    print(f"Author: {book.author}")
    print(f"Language: {book.language}")
    print(f"Category: {book.category}")
    print(f"Average Rate: {book.average_rate}")
    print(f"Maturity: {book.maturity}")
    print(f"Publisher: {book.publisher}")
    print(f"Published Date: {book.published_date}")
    print(f"Page Count: {book.page_count}")
    print("----------------------")
```

```python
#this function lets the user get 100 books randomly , they are the books that he already read
def user100(user):
    """user100(user): This function adds 100 randomly selected books to the user's reading list and removes them from the main books list. This ensures that the user is recommended books they haven't read yet. From this point, we begin analyzing the user's books to identify their preferences."""
    for i in range(100):
        book=random.choice(books)
        books.remove(book)      
        user.books.append(book)
```

```python
def best_authors(user):
    """best_authors(user): Returns a dictionary of authors the user has already read books from. The values represent the number of times the user has read a book by each author, divided by 100. This allows for a comparison where an author with 18 read books is preferred over an author with only 4."""
    dic={}
    for book in user.books:
        if book.author in dic:
            dic[book.author]+=0.01
            continue
        dic[book.author]=0.01
    return dic
```

```python
def best_publishers(user):
    """best_publishers(user): Similar to best_authors, this function returns a dictionary of publishers the user has read books from."""
    dic={}
    for book in user.books:
        if book.publisher in dic :
            dic[book.publisher]+=0.01
            continue
        dic[book.publisher]=0.01
    return dic
```

```python
def best_languages(user):
    """best_languages(user): Returns a dictionary of all possible languages in the dataset, along with the frequency coefficient of each language."""
    dic={"en":0,"ar":0,"de":0,"sv":0}
    for book in user.books:
        for lang in dic.keys():
            if book.language==lang:
                dic[lang]+=0.01
    return dic
```

```python
def best_pages(user):
    """Best_pages(user): Simply returns the mean number of pages for all the books the user has read."""
    sm=0
    for book in user.books:
        sm += book.page_count
    return sm/100
```

```python
def best_year(user):
    """Best_year(user): Also returns the mean publication year of the 100 books the user has read."""
    sm=0
    for book in user.books:
        sm += book.published_date
    return sm/100
```

```python
def maturity(user):
    """Maturity(user): Returns a number between 0 and 1 representing the percentage of mature books in the user's total number of books."""
    sm=0
    for book in user.books:
        sm += book.maturity
    return sm/100
```

```python
genre_dict = {
    'Juvenile Fiction': ['Juvenile Nonfiction', 'Study Aids', 'Children\'s art', 'Children with mental disabilities'],
    'Medical': ['Medicine', 'Dentistry', 'Communicable diseases', 'Public health'],
    'Boston': ['Historic buildings', 'Local history'],
    'Business & Economics': ['Business & Economics surveys', 'Advertising', 'Economic surveys', 'Manufacturing industries', 'Antiques & Collectibles', 'Mass media and sports', 'Finance', 'Budget', 'Corporations'],
    'Computers': ['Computer industry', 'Electronic data processing', 'Microcomputers', 'Computer programming', 'Hardware', 'Software'],
    'Biography & Autobiography': ['Biography', 'College presidents'],
    'History': ['Great Britain', 'United States', 'Political campaigns', 'Confederate States of America', 'Maryland Campaign, 1862', 'Local history', 'Indians of North America', 'African Americans'],
    'Psychology': ['Psychology', 'Child psychology', 'Bereavement', 'Love', 'Self-Help', 'Body, Mind & Spirit', 'Feminist criticism'],
    'Law': ['Copyright', 'Lawyers', 'Charter schools', 'Legal research', 'Administrative law', 'Constitutional law', 'Civil rights', 'Civil procedure', 'Legislation'],
    'Science': ['Geology', 'Gases', 'Heat', 'Physics', 'Chemistry', 'Biology', 'Mathematics'],
    'Literary Criticism': ['Literary Criticism', 'English literature'],
    'Reference': ['Reference books', 'Subject headings, Library of Congress', 'Reference'],
    'Language Arts & Disciplines': ['Language Arts & Disciplines', 'Foreign Language Study', 'Arabic alphabet', 'Arabic literature']
}

def best_categories(user):
    genre_vectors = {}

    for book in user.books:
        category = book.category
        for genre in genre_dict:
            if category in genre_dict[genre]:
                if genre not in genre_vectors:
                    genre_vectors[genre] = 0
                genre_vectors[genre] += 0.07

    return genre_vectors

```

```python
abs=lambda x : x if x>=0 else -x
"""WEIGHTS
self.author: 8
self.language: 3
self.category: 9
self.average_rate: 10 as its between 0 and 5 we ll make it on a 0,1 scale 3/5 rate is 0.7 added score
self.maturity: 6
self.publisher: 4
self.published_date: 2
self.page_count: 5"""

def score(book,user):
    """A lambda function returns the absolute value of a number, whether it's an integer or a float.

The commenter shows the weights chosen for each attribute.

Score(book, user): This function determines the compatibility between a book and a user based on the user's book list. It starts with authors, publishers, and languages, adding their respective values multiplied by the weight. The book's average rating has a weight of 10, which is scaled from 0 to 0.1. Page count and published date contribute negatively to the score, with points deducted based on their difference from the user's average. Finally, the function considers the book's categories and adds their values from the categories dictionary, multiplied by a coefficient of 9."""
    score=0
    
    authors = best_authors(user)
    if book.author in authors and book.author != "Unknown" :
        score += authors[book.author]*8
    
    publishers = best_publishers(user)
    if book.publisher in publishers and book.publisher != "Unknown" :
        score += publishers[book.publisher]*4
    
    score+=best_languages(user)[book.language]*3
    
    score+=book.average_rate/50
    
    score-=abs(book.page_count-best_pages(user))*5/4284
    
    score-=abs(book.published_date-best_year(user))/900
    
    score-=abs(book.maturity-maturity(user))*6/10
    
    #category
    genres = best_categories(user)
    if book.category!="None":
        for surgenre in genre_dict:
            for subgenre in surgenre:
                if subgenre == book.category:
                    score += genres[surgenre]*9
                    break
            break
    return score
```

```python
class Node:
    """The Node class is defined, containing a data attribute and a pointer to the next object in the future."""
    def __init__(self, data):
        self.data = data
        self.next = None


class LinkedList:
    def __init__(self):
        """The LinkedList class is defined, with the head as an attribute."""
        self.head = None

    def append(self, data):
        """The append method adds an element to the tail or end of the linked list (O(n))."""
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = new_node

    def __iter__(self):
        current = self.head
        while current:
            yield current.data
            current = current.next
    def sort_by_second_element(self):
        """sort_by_second_element: Creates the linked list and sorts it by comparing each element's score. Elements are switched if the score of the current element is lower than the next element."""
        if self.head is None or self.head.next is None:
            return

        is_sorted = False
        while not is_sorted:
            is_sorted = True
            current = self.head
            while current.next:
                if current.data[1] < current.next.data[1]:
                    current.data, current.next.data = current.next.data, current.data
                    is_sorted = False
                current = current.next
    def print_elements(self):
        """print_elements: a method to print the top 10 books' names with the best accuracy score"""
        current = self.head
        i = 0
        print("Recommended books for you :\n")
        while current and i < 10:
            print('  -',current.data[0])
            current = current.next
            i+=1


```
